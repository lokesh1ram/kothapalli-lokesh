```python
#import the libraries
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# import .csv file and convert it to a Dataframe object
df = pd.read_csv(r"C:\Users\DELL\Downloads\Wholesale customers data.csv")
df.head()
```

```python
print(df.info())
```

```python
df.drop(['Channel','Region'],axis=1,inplace=True)
```

```python
array=df.values
```

```python
array
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array)
```

```python
X
```

```python
import scipy.cluster.hierarchy as sch
```

```python
plt.figure(figsize=(20,6))
dendo = sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('Dendrogram')
plt.xlabel('Customer data')
plt.ylabel('Eucl Distance')
plt.show()
```

```python
len(set(dendo['color_list']))-1
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
group = AgglomerativeClustering(n_clusters=3)
group.fit_predict(X)
```

```python
cluster = group.fit_predict(X)
cluster.shape
```

```python
# Model evaluation tool
from sklearn.metrics import silhouette_score
```

```python
# Check how well your model is making cluster
silhouette_score(X,cluster)
```

```python
X
```

```python
from sklearn.cluster import KMeans
import numpy as np

# Assuming X is already defined as a NumPy array or DataFrame
wcss = []
for i in range(2, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
plt.plot(range(2,11),wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WSCC')
plt.show()
```

```python
#fitting k-means to the dataset
model = KMeans(n_clusters = 4, random_state = 309)
groups = model.fit_predict(X)
```

```python
type(groups)
```

```python
group_num=pd.DataFrame(groups,columns=['Group'])
group_num
```

```python
groups
```

```python
cust_kmeans_data=pd.concat([df,group_num],axis=1)
cust_kmeans_data
```

```python
silhouette_score(X,groups)
```
