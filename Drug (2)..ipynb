```python
top_conditions = df["Condition"].value_counts().nlargest(10)
sns.barplot(x=top_conditions.values, y=top_conditions.index)
```

```python

plt.figure(figsize=(12, 6))
plt.subplot(1, 3, 1)
sns.histplot(df['Effective'], bins=20, kde=True)
plt.title('Effectiveness Distribution')


plt.subplot(1, 3, 2)
sns.histplot(df['EaseOfUse'], bins=20, kde=True)
plt.title('Ease of Use Distribution')

plt.subplot(1, 3, 3)
sns.histplot(df['Satisfaction'], bins=20, kde=True)
plt.title('Satisfaction Distribution')

plt.tight_layout()
plt.show()
```

```python

numerical_df = df.select_dtypes(include=['float64', 'int64'])


corr_matrix = numerical_df.corr()


plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.boxplot(x='Condition', y='Satisfaction', data=df)
plt.xticks(rotation=90)
plt.title('Satisfaction by Medical Condition')
plt.show()


plt.figure(figsize=(8, 6))
sns.boxplot(x='Type', y='Satisfaction', data=df)
plt.title('Satisfaction by Drug Type')
plt.show()
```

```python

plt.figure(figsize=(8, 6))
sns.scatterplot(x='Reviews', y='Satisfaction', data=df)
plt.title('Reviews vs Satisfaction')
plt.show()
```

```python

df_encoded = pd.get_dummies(df, columns=['Condition', 'Drug', 'Indication', 'Type'], drop_first=True)

df_encoded.head()
```

```python

plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['Effective', 'EaseOfUse', 'Satisfaction']])
plt.title('Outliers Detection')
plt.show()


Q1 = df[['Effective', 'EaseOfUse', 'Satisfaction']].quantile(0.25)
Q3 = df[['Effective', 'EaseOfUse', 'Satisfaction']].quantile(0.75)
IQR = Q3 - Q1
outliers = ((df[['Effective', 'EaseOfUse', 'Satisfaction']] < (Q1 - 1.5 * IQR)) | 
            (df[['Effective', 'EaseOfUse', 'Satisfaction']] > (Q3 + 1.5 * IQR)))


outliers.sum()
```
